```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('College_Data')
```

```python
## create a duplicate df for comparison later
dataCompare = pd.read_csv('College_Data')
dataCompare.columns = ['Name of College', 'Private', 'Apps', 'Accept', 'Enroll', 'Top10perc',
       'Top25perc', 'F.Undergrad', 'P.Undergrad', 'Outstate', 'Room.Board',
       'Books', 'Personal', 'PhD', 'Terminal', 'S.F.Ratio', 'perc.alumni',
       'Expend', 'Grad.Rate']
dataCompare.head()
```

```python
##Create a seperate df for Private
isPrivate = data['Private']
isPrivate.head(2)
```

```python
## function to convert Yes/No to bool
def priPub(x):
    if (x=='Yes'):
        return 1
    else:
        return 0
```

```python
isPrivateBin = isPrivate.apply(priPub)
isPrivateBin = isPrivateBin.reset_index()
isPrivateBinDf = pd.DataFrame(isPrivateBin)
isPrivateBinDf.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
isPrivateBinDf.drop('index', axis=1, inplace=True)
isPrivateBinDf.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2, verbose=10, random_state=200)
```

```python
kmeans.fit(data)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
len(kmeans.labels_)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(isPrivateBinDf, kmeans.labels_)
```

```python
print(classification_report(isPrivateBinDf, kmeans.labels_))
```

```python
##Pretty good precision considering algorithm uses just the features to cluster 
##the universities. 
```

```python
convertedLabels = kmeans.labels_
convertedLabels
```

```python
range(len(convertedLabels))
```

```python
L = []
for x in range(len(convertedLabels)-1):
    L.append(str(convertedLabels[x]))
```

```python
for x in range(len(convertedLabels)-1):
    if(L[x]=='1'):
        L[x] = 'Private'
    else:
        L[x] = 'Public'
```

```python
resultsBin = pd.DataFrame(L, columns=['Type Predicted'])
resultsBin.head(2)
```

```python
predictionsDf = pd.concat([dataCompare, resultsBin], axis=1)
predictionsDf.head(2)
```

```python
predictionsDf.to_csv('predictions_CollegeType.csv', header=True, index_label='Index')
```
